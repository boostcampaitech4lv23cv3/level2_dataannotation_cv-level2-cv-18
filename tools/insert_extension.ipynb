# hiertext일 경우 .jpg 파일 삽입하기

- 현재 HierText_Convert_UFO.ipynb는 파일 이름만 불러오고 '.jpg' 확장자를 붙이지 않음
- json을 load하여 그냥 사용할 경우에 파일을 불러올 수 없어서 오류가 발생
- 이를 해결하기 위한 .jpg 삽입 코드

```python
import json
import os
```

```python
data_dir = '/opt/ml/input/data/hiertext'  ## datasets path

save_path = '/opt/ml/input/data/hiertext'

with open(os.path.join(data_dir, 'ufo/{}.json'.format('train')), 'r') as f:
    anno = json.load(f)
    
image_fnames = sorted(anno['images'].keys())
print(len(image_fnames))
```

```python
for file in image_fnames:
    anno['images'][f'{file}.jpg'] = anno['images'][file]
    del anno['images'][file]

print(len(anno['images'].keys()))
print(anno['images'].keys())
```

```python
with open(os.path.join(save_path,'ufo','train.json'), 'w') as outfile:
    json.dump(anno, outfile)
```

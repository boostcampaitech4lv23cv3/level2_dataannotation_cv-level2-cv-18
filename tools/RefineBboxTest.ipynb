# Refine BBox Test

```python
import os
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../')
from dataset import SceneTextDataset
```

```python
CONFIG = {
    'PATH_ROOT' : '../../input/data/merged-dataset/',
    'SPLIT_KEY' : 'train'
}
```

```python
dataset = SceneTextDataset(CONFIG['PATH_ROOT'], CONFIG['SPLIT_KEY'])
number_of_data = len(dataset)
target_index = 61
image, word_bboxes, roi_mask = dataset[target_index]
def drawBBox(image:np.ndarray, word_bboxes:np.ndarray):
    word_bboxes = np.clip(word_bboxes,0,image.shape[0])
    contours = []
    for bbox in word_bboxes:
        contours.append(np.int32(bbox))
    image = image * 127.0
    image = image + 127.0
    image = np.array(image, np.ubyte)
    image = cv2.drawContours(image, contours, -1, (255,255,0))
    return image
plt.imshow(drawBBox(image=image, word_bboxes=word_bboxes))
```
